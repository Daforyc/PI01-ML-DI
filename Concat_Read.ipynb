```python
import numpy as np
import pandas as pd
```

```python
#creo funcion para modificar los nombres del ID
def namechange(a, x):
        if a == 1: 
                return "a"+x
        if a == 2: 
                return "d"+x
        if a == 3: 
                return "h"+x
        if a == 4: 
                return "n"+x
```

```python
#Leo los CSV de peliculas
am_df =pd.read_csv('Datasets/amazon_prime_titles.csv')
dy_df=pd.read_csv('Datasets/disney_plus_titles.csv')
hu_df = pd.read_csv('Datasets/hulu_titles.csv')
nfx_df = pd.read_csv('Datasets/netflix_titles.csv')
```

```python
#Leo los csv de ratingScore
r1_df=pd.read_csv('Datasets/1.csv')
r2_df=pd.read_csv('Datasets/2.csv')
r3_df=pd.read_csv('Datasets/3.csv')
r4_df=pd.read_csv('Datasets/4.csv')
r5_df=pd.read_csv('Datasets/5.csv')
r6_df=pd.read_csv('Datasets/6.csv')
r7_df=pd.read_csv('Datasets/7.csv')
r8_df=pd.read_csv('Datasets/8.csv')
```

```python
am_df['show_id'] = am_df['show_id'].apply(lambda x: namechange(1,x))
am_df.head()
```

```python
hu_df['show_id'] = hu_df['show_id'].apply(lambda x: namechange(3,x))
hu_df.head()
```

```python
nfx_df.head()
```

```python
nfx_df.info()
nfx_df.isnull().sum()
```

```python
nfx_df['show_id'] = nfx_df['show_id'].apply(lambda x: namechange(4,x))
nfx_df.head()
```

```python
#Despues de hacer los cambios en los codigos concateno los csv y genero uno general
gen_df = pd.concat([am_df, dy_df, hu_df, nfx_df])
gen_df.head()
```

```python
#remplazo valores nulos del campo rating por una g
gen_df["rating"].fillna(value ="g", inplace = True)
```

```python
#Conversión de date_added al formato adecuado (AAAA-mm-dd)
gen_df["date_added"] = pd.to_datetime(gen_df.date_added)
```

```python
#Conversión de campos de texto a minuscula utilizando el metodo "applymap"
gen_df = gen_df.applymap(lambda s: s.lower() if type(s) == str else s)
```

```python
#Separación de columna "duration" en dos ("duration_int" y "duration_type")
gen_df["duration_int"] = gen_df["duration"].str.split(" ").str.get(0)
gen_df["duration_type"] = gen_df["duration"].str.split(" ").str.get(1)
```

```python
#En nueva columna "duration_type" unificar season y seasons en "season":
gen_df["duration_type"] = gen_df["duration_type"].replace(["Seasons", "seasons"], "season")
```
